```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris
```

```python
from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2, 
                                                    random_state=13)

X_train.shape, X_test.shape
```

```python
import numpy as np

# 중복제거
np.unique(y_train, return_counts=True)
```

```python
# 규제, LASSO 란 나의 train데이터에 과적합되지 않도록 제한을 걸어두는 것!

from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2,
                                                    stratify=labels, 
                                                    random_state=13)

X_train.shape, X_test.shape

```

```python
import numpy as np

np.unique(y_test, return_counts = True)
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(18, 12))
plot_tree(iris_tree, filled=True, rounded=True, feature_names=iris.feature_names[2:], class_names=iris.target_names)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(12, 8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()


y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
scatter_highlight_kwargs = {'s':150, 'edgecolor':'Red', 'label':'Test data', 'alpha':0.7} # 150개 데이터 전체 
scatter_kwargs = {'s':120, 'edgecolor':'#FFFFFF', 'alpha':0.9} # scatter_highlight 적용

plt.figure(figsize=(14,10))
plot_decision_regions(X=features, y=labels,
                      X_highlight=X_test, clf=iris_tree, legend=2,
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs=scatter_kwargs,
                      contourf_kwargs={'alpha':0.2})
plt.show()
```

```python
features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size=0.2,
                                                    stratify=labels,
                                                    random_state=13)
iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(18, 12))
plot_tree(iris_tree, filled=True, rounded=True, feature_names=iris.feature_names, class_names=iris.target_names)
```

```python
# test_data를 예측했을 때, 1번으로 예측
test_data = np.array([[4.3, 2., 1.2, 1.0]])
iris_tree.predict(test_data)
```

```python
# 주요 특성 확인? max depth가 2이므로 2개의 특성이 나옴
iris_tree.feature_importances_
```

```python
# max depth가 2이므로 2개의 특성 중요도 나옴
iris_clf_model = dict(zip(iris.feature_names, iris_tree.feature_importances_))
iris_clf_model
```
